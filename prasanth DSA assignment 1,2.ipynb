```python
# 1. checking a number even or odd

#step1 declare a variable that has to store input as integer number

x = int(input("Enter a number:"))

#step2 to check the number is even or odd we need to write condition, here iam using if condition

if x % 2 == 0:                 # % implies modulus operator, gives remainder of division  
    print(x, "is even number")
else:
    print(x, " is odd number")
```

```python
# 2. checking a number whether it is positive, negative or zero

#step1 declare a variable that has to store input number

num = float(input("Enter a number:")) # declaring input number as float

# step2 using if elif condition for determining a number is +, - or 0

if num == 0:
    print(num, "is zero")
elif num > 0:
    print(num, "is positive number")
else:
    print(num, "is negative number")

```

```python
#3. checking year is a leap year or not using for loop
# intializing variables

x = int(input("Enter the year: "))
y = int(input("how many years to calculate: "))

# for loop with range function

for i in range(0,y+1):                                # y+1 implies index + 1 
    z = x+i                                           # year incrementation 
    if z % 4 == 0 and (z % 100 != 0 or z % 400 == 0): # if condition to calculate leap year
        print(z, "is a leap year")
    else:
        print(z, "not a leap year")
```

```python
#5. performing the operations on two numbers using different operators like +,-,* & / using elif 

# step1 intialising input numbers
x = int(input("Enter the first number: "))
y = int(input("Enter the second number: "))

#step2 giving a choice of operations has to perform on numbers
z = input(" Enter the one of the particular operation +,-,* & /:  ")

#step3 writing conditional statements

if z == '+':
    output = x + y
elif z == '-':
    output = x - y
elif z == '*':
    output = x * y
elif z == '/':
    output = x / y
    

else:
    #print("input operation not found")
    output = "invalid"
print(x, z, y, ":", output)

```

```python
#6 reversing a number
# Ask for enter the number from the use  
x = int(input("Enter the number: "))  
  
# Initiate value to null  
reverse = 0  
  
# reverse the integer number using the while loop  
  
while (x > 0):  
    # Logic  
    remainder = x % 10  
    reverse = (reverse * 10) + remainder  # for suppose number is 123 we do 123 % 10 = 3 intial 0*10+3=3 next 3*10+ remainder=30+2=32
    x = x // 10                           # 12//Now loop will iterate on this number
  
# Display the result

print("The reverse number is : {}".format(reverse))  
```

```python
#9 calculating electricity bill according to the unit price
#step1 giving input variable

units=int(input("enter no of units: "))

# using elif condition
if units<=50:
    result= units*0.5
    print("total electricity bill: ",result)
elif units <= 150:
    result= (units-50)*0.75+ (50*0.5)
    print("total electricity bill: ",result)
elif units <= 250:
    result= (50*0.5)+(150-50)*0.75+(units-150)*1.2
    print("total electricity bill: ",result)
elif units > 250:
    result= (50*0.5)+(150-50)*0.75+(250-150)*1.2 + (units-250)*1.5
    print("total electricity bill: ",result)
else:
    print("no bill")
```

```python
# 1.program to add all elements of the array
#step1 importing array module
import array  
def add(array): # we define a function by add as function name 
    step = 0
    for i in array:
        step = step + i
    return(step)
array = array.array('i', [5,6,-7,8]) #array class of the module that helps us to create array, 'i' signed integer with maximum size in bytes 2
res = add(array) # call the function
print("addtion result of array: ",res)

```
